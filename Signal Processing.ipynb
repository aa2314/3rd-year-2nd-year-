```python
import numpy as np
import matplotlib.pyplot as plt
g = np.random.randn(1000)
x_g = np.linspace( - 3 * 1,  + 3 * 1, 1000)
y_g =[(1 / (np.sqrt(2 * np.pi) * abs(1))) * np.exp(- ((x) / abs(1)) **2 / 2) for x in x_g]

plt.hist(g, bins = 20, density=True)
plt.plot(x_g, y_g)
plt.show()
u = np.random.rand(1000)
plt.hist(u, bins = 20, density = True)
x_u = np.linspace(0, 1, 1000)
plt.plot(x_u, 1000* [1])
plt.show()
def ksdensity(data, width=0.3):
    """Returns kernel smoothing function from data points in data"""
    def ksd(x_axis):
        def n_pdf(x, mu=0., sigma=1.): # normal pdf
            u = (x - mu) / abs(sigma)
            y = (1 / (np.sqrt(2 * np.pi) * abs(sigma)))
            y *= np.exp(-u * u / 2)
            return y
        prob = [n_pdf(x_i, data, width) for x_i in x_axis]
        pdf = [np.average(pr) for pr in prob] # each row is one x value
        return np.array(pdf)
    return ksd
# Plot normal distribution
fig, ax = plt.subplots(2)
x = np.random.randn(1000)
ax[0].hist(x, bins=30) # number of bins
ks_density = ksdensity(x, width=0.4)
# np.linspace(start, stop, number of steps)
x_values = np.linspace(-5., 5., 100)
ax[1].plot(x_values, ks_density(x_values))
# Plot uniform distribution
fig2, ax2 = plt.subplots(2)
x = np.random.rand(1000)
ax2[0].hist(x, bins=20)
ks_density = ksdensity(x, width=0.2)
x_values = np.linspace(-1., 2., 100)
ax2[1].plot(x_values, ks_density(x_values))
```

```python
a = 4
b = 10
x = np.random.randn(1000)
y1 = a * x + b
x_j1 = np.linspace(b - 3 * a, b + 3 * a, 1000)
y_j1 =[(1 / (np.sqrt(2 * np.pi) * abs(a))) * np.exp(- ((x - b) / abs(a)) **2 / 2) for x in x_j1]
plt.hist(y1, bins = 20, density = True)
plt.plot(x_j1, y_j1, label = " y = ax+b")
plt.legend()
plt.show()
y2 = x ** 2
x_j2 = np.linspace(0.01, 10, 1000) ## should be careful with x = 0 here
y_j2 = [(np.exp(- x/2)  / (np.sqrt(2 * np.pi * x))) for x in x_j2]
plt.hist(y2, bins = 50, density = True)
plt.plot(x_j2, y_j2, label = "y = x^2")
plt.legend()
plt.show()
```

```python
n = 1000
V = np.random.exponential(1, n)
U = (np.random.rand(n) - 1/2) * np.pi
def f(alpha, beta):
    b = (1/alpha) * np.arctan(beta * np.tan( np.pi * alpha / 2))
    s = (1 + beta**2 * (np.tan(np.pi * alpha / 2))** 2) ** (1/(2* alpha))
    X  = s * (np.sin(alpha*(U + b))/(np.cos(U))** (1/alpha)) * ((np.cos(U - alpha* (U + b)))/V)** ((1-alpha)/alpha)
    return X
```
